Question 1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?


Answer:

In hypothesis testing, we begin with a null hypothesis (Hâ‚€) and an alternative hypothesis (Hâ‚). Errors can occur because decisions are based on sample data rather than the entire population.

**Type I Error (False Positive)**

->Occurs when Hâ‚€ is true, but we incorrectly reject it.

->Denoted by Î± (alpha).

Example: Concluding that a medicine works when it actually does NOT work.

Impact: Leads to adopting a wrong idea or treatment.

**Type II Error (False Negative)**

->Occurs when Hâ‚€ is false, but we fail to reject it.

->Denoted by Î² (beta).

Example: Concluding that a medicine does NOT work when it actually works.

Impact: Misses a real effect; potential benefits are lost.

**Impact on Decision-Making**

->Lowering Type I error increases Type II error and vice-versa â€” a balance is needed.

->Critical for fields like medicine, finance, and manufacturing, where wrong decisions have major consequences.

**Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?**


Answer:


Purpose of ANOVA:

ANOVA (Analysis of Variance) tests whether three or more groups have significantly different means.

Hypotheses:

->
Hâ‚€: All group means are equal

->
Hâ‚: At least one group mean is different

Assumptions:

1  Normal distribution of groups

2 Equal variances (homogeneity of variance)

3 Independent observations

Extension Beyond Two Groups:

->
A t-test only compares two means.

-> ANOVA compares multiple groups simultaneously, reducing overall Type I error.

```python
#Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
#Answer:
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample dataset
data = [12, 10, 13, 15, 14, 11, 16, 14, 13, 12]

# Population mean (hypothesized)
mu_0 = 12

# Perform one-sample Z-test
z_stat, p_value = ztest(data, value=mu_0)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to Reject H0: No significant difference from the population mean.")
```

```python
'''
Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
ANSWER
'''

import numpy as np
import matplotlib.pyplot as plt

# Non-normal distribution (exponential)
population = np.random.exponential(scale=1, size=100000)

sample_means = []

# Draw 1000 samples of size 50 each
for _ in range(1000):
    sample = np.random.choice(population, size=50)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30)
plt.title("Sampling Distribution of Means (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.
#Answer:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sd = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    margin = z * sd / np.sqrt(n)

    lower = mean - margin
    upper = mean + margin

    print("Mean:", mean)
    print("Confidence Interval:", (lower, upper))

    # Plot
    plt.errorbar(1, mean, yerr=margin, fmt='o')
    plt.title("Confidence Interval for Mean")
    plt.xlim(0, 2)
    plt.xticks([])
    plt.show()

# Example usage
data = [12, 10, 13, 15, 14, 11, 16]
plot_confidence_interval(data)
```

```python
#Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
#Answer:
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([30, 20, 25, 25])

# Expected frequencies
expected = np.array([25, 25, 25, 25])

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0: Observed distribution is significantly different from expected.")
else:
    print("Fail to Reject H0: No significant difference between observed and expected.")
```
